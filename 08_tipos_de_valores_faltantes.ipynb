## Tipos de valores faltantes

```python
import pandas as pd
import numpy as np
import missing
import missingno
```

```python
pima_diabetes_df = pd.read_csv("./data/pima-indians-diabetes.csv")
pima_diabetes_df.head(2)
```

```python
(
    pima_diabetes_df
    .missing
    .missing_variable_plot()
)
```

```python
(
    pima_diabetes_df[pima_diabetes_df.columns[1:6]]
    .replace(0, np.nan)
    .missing.missing_variable_plot()
)
```

### *Missing completely at Random* (MCAR)

```python
(
    pima_diabetes_df
    .replace(0, np.nan)
    .missing.sort_variables_by_missingness()
    .pipe(
        missingno.matrix
    )
)
```

### *Missing at Random* (MAR)

```python
(
    pima_diabetes_df
    .replace(0, np.nan)
    .missing.sort_variables_by_missingness()
    .sort_values(by="blood_pressure")
    .pipe(
        missingno.matrix
    )
)
```

### *Missing Not At Random* (MNAR)

```python
(
    pima_diabetes_df
    .replace(0, np.nan)
    .missing.sort_variables_by_missingness()
    .sort_values(by="insulin")
    .pipe(
        missingno.matrix
    )
)
```
