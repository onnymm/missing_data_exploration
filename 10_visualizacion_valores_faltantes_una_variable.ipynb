## Visualización de valores faltantes en una sola variable

```python
import pandas as pd
import numpy as np
import missing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
riskfactors = pd.read_csv("./data/riskfactors.csv")
riskfactors.head(2)
```

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
    .pipe(
        lambda df: (
            sns.boxenplot(
                data= df,
                x= 'weight_lbs_NA',
                y= 'age',
                hue= 'weight_lbs_NA'
            )
        )
    )
)
```

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
    .pipe(
        lambda df: (
            sns.displot(
                data= df,
                x= 'age',
                hue= 'weight_lbs_NA',
                kind= 'kde'
            )
        )
    )
)
```

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
    .pipe(
        lambda df: (
            sns.displot(
                data= df,
                x= 'age',
                col= 'weight_lbs_NA',
                facet_kws= {
                    'sharey': False
                },
                bins= 30
            )
        )
    )
)
```

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
    .pipe(
        lambda df: (
            sns.displot(
                data= df,
                x= 'age',
                col= "marital_NA",
                row= 'weight_lbs_NA'
            )
        )
    )
)
```
