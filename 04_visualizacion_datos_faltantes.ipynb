```python
import pandas as pd
import missingno
```

```python
riskfactors = pd.read_csv("data/riskfactors.csv")
riskfactors.head(5)
```

## Visualización inicial de valores faltantes

### Variable

```python
riskfactors.missing.missing_variable_plot()
```

### Casos / Observaciones / Filas

```python
riskfactors.missing.missing_case_plot()
```

```python
(
    riskfactors
    .missing
    .missing_variable_span_plot(
        variable= "weight_lbs",
        span_every= 10,
        rot= 0
    )
)
```

```python
missingno.bar(df= riskfactors)
```

```python
missingno.matrix(df= riskfactors)
```

```python
(
    riskfactors
    .missing
    .missing_upsetplot(
        variables= ["pregnant", "weight_lbs", "smoke_stop"],
        element_size= 60
    )
)
```
