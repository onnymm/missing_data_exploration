```python
import pandas as pd
import numpy as np
import missing
```

```python
riskfactors = pd.read_csv("./data/riskfactors.csv")
riskfactors.head(2)
```

## Matriz de sombras

### Concepto y aplicación de la matriz de sombras

Construcción de la matriz de sombras

```python
(
    riskfactors
    .isna()
    .replace(
        {
            True: "Missing",
            False: "Not missing"
        }
    )
    .add_suffix("_NA")
    .pipe(
        lambda df: (
            pd.concat(
                [riskfactors, df],
                axis= "columns"
            )
        )
    )
)
```

### Utilizar función de utilería `bind_shadow_matrix()`

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
)
```

Explorar estadísticos utilizando las nuevas columnas de la matriz de sombras

```python
(
    riskfactors
    .missing.bind_shadow_matrix(only_missing= True)
    .groupby(["weight_lbs_NA"])
    ["age"]
    .describe()
)
```
